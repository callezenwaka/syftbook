#### 2.1 Purpose of Creating Mock Data

Generating mock data in AI development offers several advantages in different stages of the development process. One of the major advantages is realized in the privacy-preserving AI landscape. Some of the key benefits in this area are listed below. 

- **Staging Environment for Data Scientists:** A mock dataset offers the data scientists a method to explore the data and simulate their intended study in a realistic way without exposing sensitive user information. Moreover, as the mock dataset replicates the most important attributes of the private dataset, it can be used for testing that the code works as intended before submitting for remote code execution, as a staging environment that minimizes friction for conducting a study on data researchers don't have access to.
<br>

- **Ensuring Complete Partisan of Sensitive Data:** When dealing with actual user data, especially private information, it introduces several critical considerations related to privacy, security, and intellectual property. Additionally, complying with legal regulations such as the GDPR can create significant challenges for all parties involved. To address these concerns, data owners can turn to the use of mock datasets to create a controlled testing environment. Unlike working with real data, mock datasets come with minimal risks and offer data scientists the freedom to experiment without exposing private information. The actual computation on private data only occurs and results are only revealed after a thorough review by the data owners, ensuring that data misuse is prevented.

    This approach provides data owners with the assurance that researchers only have access to simulated data for experimentation purposes, eliminating the risk of unauthorized access to sensitive information. In the safest PySyft paradigms, the mock and private datasets are stored on separate machines, creating an additional layer of security. This separation ensures that even potentially malicious researchers cannot bypass the firewall and gain access to the private data. As a result, such separation establishes a secure collaborative space that benefits both data owners and data scientists or researchers.
    <br>

- **Data Control:** With mock data, Data Owners have full control over the data generation process. They can easily customize it to suit specific testing or development needs, limit certain use-cases, and prevent information leakage using linkage attacks, which is not always possible with real data.
<br>

Apart from the above mentioned advantages, in general, creating mock data can also help Data Owners create a wide range of data scenarios and edge cases that might be challenging to include in real-world data, but expected to be encountered occassionally in production environment. This enables the data scientists reduce bias in AI models while training, and testing the robustness and adaptability of the models.

#### 2.2 Methods for Creating Mock Data

There are several methods available for generating mock data, which serves as a replacement for real datasets. These methods can be ranked in terms of their potential risks of privacy leaks:

1. **Subset Extraction from Real Private Data:** The first method involves selecting a small portion of actual private data. This approach, while useful, should be employed cautiously, as it can still pose privacy risks.
<br>

2. **Anonymization of Real Private Data:** The second method is the anonymization of either the entire dataset or a subset of it. Anonymization is a process that removes or masks personally identifiable information (PII). It's a commonly used technique, but it's essential to be aware that even anonymized data can sometimes carry privacy risks.
<br>

3. **Programmatically Generated Fully Fake Data:** The third approach involves creating entirely synthetic data programmatically. This method eliminates the need for real data altogether and eliminates any potential privacy concerns associated with real data.
<br>

In many cases, it is sufficient and common to opt for either approach 1 or 2, especially when the sensitivity of the data and the risks associated with potential data leakage are relatively low. However, in situations where even sharing a small subset of data, even in an anonymized manner, could pose significant privacy risks, it is advisable to leverage approach 3 by generating entirely synthetic data to ensure maximum privacy protection.

For approach 3, data owners can take leverage of existing Python libraries and tools available for generating mock data mimicking real data. These libraries can help them create synthetic datasets that mimic the characteristics of real data. Some popular ones are listed below.

- [Faker](https://faker.readthedocs.io/en/master/index.html): Faker is a Python library that generates fake data for various data types, such as names, addresses, dates, and more. It's widely used for creating mock data for testing and development purposes.
<br>

- [Mimesis](https://mimesis.name/en/master/): Mimesis is another Python library for generating mock data. It provides support for various data types and includes localization for different countries and languages. This tool is useful for populating testing databases, creating fake API endpoints, generating custom structures in JSON and XML files, and anonymizing production data, among other things.
<br>

- [Pydbgen](https://github.com/tirthajyoti/pydbgen): Pydbgen is a Python library for generating random, fake data in bulk with support for various data types, including names, addresses, dates, and more. This Python package generates a random database `TABLE` (or a Pandas dataframe, or an Excel file) based on user's choice of data types, add `PRIMARY KEY` and insert it to new or existing database.
<br>

- [FakerJS](https://fakerjs.dev/): FakerJS is a JavaScript library for generating massive amounts of fake data. It can generate category-specific data including Product, Finance, Locations, Time-based data etc. It can also be used in combination with Python via a tool like Node.js to create mock data for web applications and APIs. 
<br>

- [Generative AI](https://github.com/steven2358/awesome-generative-ai): Generative AI models are one of the most advanced additions to the machine learning algorithm suites. These models are trained on huge corpus of data from the web and have gained the capability to generate unique, novel outputs in the forms of texts, images, audio etc. in a way that is indistinguishable from their human generated counterparts. Among many applications, these tools are great candidates for generating extremely realistic mock data.

#### 2.3 Rule of Thumbs for Creating Mock Data

While creating mock data, it is essential to ensure that the generated data represents the real data as closely as possible without violating privacy and security of the data subjects. Here are some rule of thumbs to make sure that your mock data closely mirrors real data:

- **Understand the Data Domain**:
Gain a deep understanding of the domain and context of the real data you're trying to mimic. This includes understanding the data types, relationships, and business rules associated with the data.
<br>

- **Use Realistic Data Distributions**:
Analyze the statistical properties of the real data, such as `mean`, `variance`, `min`, `max` and other relevant `distribution`. Then generate mock data with similar statistical characteristics to reflect the variability of the real data. Below are some example of questions data owners should ask themselves while generating a mock data. <br>

    - Are the individual distributions of each variable a realistic representation of the actual data distributions?
    - Have we correctly captured the base rates for categorical values?
    - Have we addressed any class imbalances that may exist in the real data?
    - Do the copula or joint distributions of variable pairs accurately mimic the relationships found in the real data distribution?
<br>


- **Maintain Data Consistency**:
Ensure that the relationships between different data attributes are preserved. For example, if you are generating height and weight for data subjects with genders male and female, you should make sure that the average height and weight of each gender matches with the average values as represented in the real data.
<br>

- **Handle Missing Data, Duplicates and Outliers**:
Real data often contains missing and/or duplicate values and outliers. Create mock data that includes these variations in a way that reflects the patterns observed in the real data. Here are some key factors to keep in mind when replicating missing data and outlier distribution in mock data.<br>

    - Are outliers explicitly modeled in the mock data?
    - Are missing / invalid / exceptional values modeled, and are their distributions realistic? Consider using the appropriate missing data distribution approach such as missing completely at random (MCAR), missing at random (MAR) and missing not at random (MNAR) etc.
    
<br>

- **Respect Data Constraints**:
Real data may have constraints and specific business rules. Respect these constraints when generating mock data. For example, if a field has a maximum length, ensure that your mock data adheres to this limit.
<br>

- **Consider Data Volume**:
Although it is not necessary that volume of the mock data matches the amount of data that exists in the real scenario, for certain use-cases it may be beneficial. Matching the data volumne will ensure that the AI models trained on them are tested on the realistic scale of the data and be able to process them within the expected time limit. Also, if the real dataset is expected to grow over time, it is a good practice to consider how the mock data and models trained on them will scale. In the cases where the mock data is only a sub sample of the real data in terms of volume, it is essential for the data owners to inform the data scientists about this. Data owners can include this information in the `asset` and `dataset` description fields in PySyft.
<br>

- **Incorporate Noise and Variability**:
Real data often contains noise and variability. Introduce appropriate level of randomness and variability into your mock data to simulate the unpredictability of real-world scenarios.
<br>

- **Use Realistic Data Sources**:
If possible, use real data sources such as dictionaries, lists of names, or location data to populate your mock data. This can make the data more realistic and diverse.
<br>

- **Validate Against Real Data**:
Compare your mock data to real data to ensure that it aligns with the actual data as closely as possible. If available, use validation techniques to measure the similarity between the two datasets.
<br>

- **Consider Privacy and Security**:
Be cautious when dealing with sensitive or personally identifiable information (PII). Anonymize or pseudonymize data as needed to protect privacy while using them as samples for the mock data.
<br>

- **Document the Generation Process**:
Document how you generated the mock data, including any assumptions or modifications you made. This documentation is crucial for transparency, debugging and reproducibility.
<br>

- **Iterate and Refine**:
Creating realistic mock data may require several iterations. Be prepared to iteratively refine the generation process if needed based on feedback from external users (i.e. data scientists) and changes in real data.
<br>

- **Involve Domain Experts**:
Collaborate with domain experts, data analysts, or stakeholders who have a deep understanding of the data. Their insights can help ensure that the mock data accurately represents the real data.
<br>

- **Monitor and Update**:
Regularly review and update your mock data to keep it in sync with changes in the real data. This is especially important for evolving systems.

**Mock Data Generation Steps:**

- Generate Fake `Id` and `Name`
- Generate Fake `Short Description`
- Encode each unique `Gender` with a standard naming convention (i.e. `Gender 1`, `Gender 2` etc.) \
to protect the under represented genders and generate Fake transformed `Gender` value

**Step 1: Process Real Data Temporarily**

```python
## convert <string> type list of strings to python <list> type
df["Associated Countries"] = df["Associated Countries"].apply(ast.literal_eval)

df["Associated Country Coordinates (Lat/Lon)"] = df[
    "Associated Country Coordinates (Lat/Lon)"
].apply(ast.literal_eval)

df["Associated Country Life Expectancy"] = df[
    "Associated Country Life Expectancy"
].apply(ast.literal_eval)
```

Because `Associated Countries`, `Associated Country Coordinates (Lat/Lon)` and `Associated Country Life Expectancy` are correlated with each other, it is important to preserve this relation in the mock dataset. To keep track of the corresponding mappings of these three columns, we should first unpack them from the lists and then create a dictionary where Countries will be `keys` and the rest two columns will be `values`. For this example, we will only keep `Associated Country Life Expectancy` and discard `Associated Country Coordinates (Lat/Lon)`.

```python
## separate countries from lists and calculate their individual value_counts() which will be
## used by random.choice function later as distributions
def value_counts_of_lists(series_with_lists):
    # Concatenate all the lists in the Series into a single list
    unpacked_list = [item for sublist in series_with_lists for item in sublist]

    # Create a new Series from the unpacked list
    unpacked_series = pd.Series(unpacked_list)

    # Use value_counts to get the count of unique values
    value_counts = unpacked_series.value_counts()

    return unpacked_list, value_counts


## create a dictionary where each unique country from all the lists in Associated Countries is
## are keys and the corresponding (Lat,Long) tuples are the values

unpacked_cnt_list, cnt_value_counts = value_counts_of_lists(
    df["Associated Countries"].values
)

unpacked_exp_list, exp_value_counts = value_counts_of_lists(
    df["Associated Country Life Expectancy"].values
)

print(len(unpacked_cnt_list))
print(len(unpacked_exp_list))

cnt_dict = dict.fromkeys(unpacked_cnt_list, None)

for i in range(len(unpacked_exp_list)):
    if cnt_dict[unpacked_cnt_list[i]] == None:
        cnt_dict[unpacked_cnt_list[i]] = unpacked_exp_list[i]
```

**Step 2: Generate Mock Data**

```python
NUM_OF_ROWS = df.shape[0]  # 10000

Faker.seed(0)
faker = Faker()
```

```python
gender_encode_dict = {
    "Male": "Gender 1",
    "Female": "Gender 2",
    "Transgender Female": "Gender 3",
    "Transgender Male": "Gender 4",
    "Intersex": "Gender 5",
    "Eunuch; Male": "Gender 6; Gender 1",
    "Transgender Female; Female": "Gender 3; Gender 2",
}


def generate_random_choice_columns(df, num):
    ## generate Id
    id_list = np.arange(1, num + 2000)
    fake_id = np.random.choice(id_list, size=num, replace=False)
    fake_id = pd.Series(fake_id).apply(lambda x: "Q" + str(x))

    ## generate Gender
    gender_dist = df["Gender"].value_counts(normalize=True)
    gender = np.random.choice(
        df["Gender"].unique().tolist(),
        size=num,
        replace=True,
        p=gender_dist,  # probability
    )
    gender = pd.Series(gender).replace(gender_encode_dict)

    ## generate Age of death, add noise by adding random int between(-5,5) to fake age
    age_of_death_dist = df["Age of death"].value_counts(normalize=True)
    age_of_death = np.random.choice(
        df["Age of death"].unique().tolist(),
        size=num,
        replace=True,
        p=age_of_death_dist,  # probability
    )
    age_of_death = (
        pd.Series(age_of_death).apply(lambda x: x + randint(-5, 5)).astype("float64")
    )

    ## generate Associated Countries
    assc_cnt_dist = df["Associated Countries"].value_counts(normalize=True)
    assc_cnt = np.random.choice(
        df["Associated Countries"].astype(str).value_counts().keys().tolist(),
        size=num,
        replace=True,
        p=assc_cnt_dist,  # probability
    )
    assc_cnt = pd.Series(assc_cnt).apply(ast.literal_eval)

    ## generate Life Expentency using the dictionary created above
    assc_life_exp = pd.Series(assc_cnt).apply(lambda x: [cnt_dict[i] for i in x])

    ##generate Manner of death
    manner_of_death_dist = df["Manner of death"].value_counts(normalize=True)
    manner_of_death = np.random.choice(
        df["Manner of death"].unique().tolist(),
        size=num,
        replace=True,
        p=manner_of_death_dist,  # probability
    )
    manner_of_death = pd.Series(manner_of_death)

    return fake_id, gender, age_of_death, assc_cnt, assc_life_exp, manner_of_death


def make_faker_data(num):
    fake_data = [
        {
            "Name": faker.name(),
            "Short description": faker.paragraph(nb_sentences=2),
            "Occupation": faker.job(),
            "Death year": float(faker.year()),
        }
        for x in range(num)
    ]

    return fake_data
```

```python
mock_df = pd.DataFrame()
(
    mock_df["Id"],
    mock_df["Gender"],
    mock_df["Age of death"],
    mock_df["Associated Countries"],
    mock_df["Associated Country Life Expectancy"],
    mock_df["Manner of death"],
) = generate_random_choice_columns(df, num=NUM_OF_ROWS)

fake_data = pd.DataFrame(make_faker_data(num=NUM_OF_ROWS))

for col in fake_data.columns.to_list():
    mock_df[col] = fake_data[col]

## generate Birth year by subtracting Age of death from Death year
mock_df["Birth year"] = mock_df["Death year"].astype(int) - mock_df[
    "Age of death"
].astype(int)

print(mock_df.shape)
mock_df.head()
```

**Step 3: Match Shapes between Real and Mock data (specific requirement for PySyft)**

While uploading real and mock data using PySyft, PySyft's input policy requires it to exactly match the `shape (no. of rows, no. of columns)` of both the real data and mock data. The details of this mechanism are explained in future tutorials. 

 In the beginning, we made sure to generate same number of fake rows of data as the real data. Now, we need to make sure that the final mock data includes all the columns of real data, even if some of them are discarded or not meant to be shared with external data scientists. To mark these "non-shared" columns, we can either fill them with a constant value or just keep them `NoneType`. 

In this example, `Country` and `Associated Country Coordinates (Lat/Lon)` columns are not meant to be shared with external data scientists. So we will tag them as `Not Available`. 

```python
mock_df["Country"] = ["Not Available"] * mock_df.shape[0]
mock_df["Associated Country Coordinates (Lat/Lon)"] = ["Not Available"] * mock_df.shape[0]
```

**2.4.2 Texts Data**

The next example will demostrate how to generate realistic mock data for linguistic or texts based dataset. In this example, we are going to use the pre-trained `GPT-2` model for generating `fake customer reviews` in english and spanish. In upcoming tutorials, we will cover how to fine-tune these base models on your custom dataset and generate mock texts in the style of the custom data.

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
torch_device = "cuda" if torch.cuda.is_available() else "cpu"
model_name = "gpt2-medium"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name, pad_token_id=tokenizer.eos_token_id
).to(torch_device)
```

```python
# Zero-Shot Approach
# Directly asking the model to generate texts in response to a cue.
def generate_reviews(prompt, product, n_texts):
    for i in range(n_texts):
        model_inputs = tokenizer(prompt, return_tensors="pt").to(torch_device)

        outputs = model.generate(
            **model_inputs,
            max_new_tokens=100,
            do_sample=True,
            top_k=10,
            top_p=0.95,
            num_return_sequences=1,
            eos_token_id=tokenizer.eos_token_id
        )

        review = tokenizer.decode(outputs[0], skip_special_tokens=True)
        product_names.append(product)
        reviews.append(review.splitlines()[0])
```

```python
products = [
    "64 inch Sony television",
    "Amazon Kindle",
    "Honda SUV car",
    "Dyson vacuum cleaner",
]
product_names = []
reviews = []

for product in products:
    # generate positive reviews
    prompt_pos = f"Recently I bought a {product}. I am happy with the purchase because"
    generate_reviews(prompt_pos, product, n_texts=2)

    # generate negative reviews
    prompt_neg = (
        f"Recently I bought a {product}. I am disappointed with the purchase because"
    )
    generate_reviews(prompt_neg, product, n_texts=2)
```

```python
mock_review_df = pd.DataFrame()
mock_review_df["product"] = product_names
mock_review_df["review"] = reviews
print(mock_review_df.shape)
mock_review_df
```

As seen from the results, sometimes these large language models are tend to provide inaccurate or hallucinated contexts. However, it does not have impact on mock data generation becuase the goal of a mock for text data is to simply provide outputs that resemble similar vocabulary, format and sequence lengths as the real data, even if the context is not always accurate.

We can also use the same approach for generating texts in any language other than english. In the following example, we are going to use the `gpt-2-spanish` model to generate customer reviews in spanish.

```python
torch_device = "cuda" if torch.cuda.is_available() else "cpu"
model_name = "datificate/gpt2-small-spanish"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# add the EOS token as PAD token to avoid warnings
model = AutoModelForCausalLM.from_pretrained(
    model_name, pad_token_id=tokenizer.eos_token_id
).to(torch_device)
```

```python
products = [
    "Televisin Sony de 64 pulgadas",
    "Amazon Kindle",
    "auto SUV Honda",
    "aspiradora Dyson",
]
product_names = []
reviews = []

for product in products:
    # generate generic reviews
    prompt = f"Ayer compr un {product}. Fue"
    generate_reviews(prompt, product, n_texts=2)
```

```python
mock_review_df = pd.DataFrame()
mock_review_df["product"] = product_names
mock_review_df["review"] = reviews
print(mock_review_df.shape)
mock_review_df
```

#### 2.4.3 Images
We can also use generative AI to create fake images based on real dataset. Similar to large language models, there are pre-trained models for images - however, images can come from subspaces that are significantly distinct from each other and a pretrained model can hardly generate images that cover the whole possible spaces (e.g. chest X-ray vs satellite images). We therefore propose using a fine-tuned generative adversarial model that addresses your use-case, which are sometimes readily available as open-source models. As an example, we will show today how we can use open-sourced fine-tuned models to generate fake chest X-ray images.
